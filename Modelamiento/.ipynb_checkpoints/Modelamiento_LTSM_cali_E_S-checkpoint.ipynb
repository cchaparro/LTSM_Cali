```python
from math import sqrt
from numpy import concatenate
from matplotlib import pyplot as plt
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import numpy as np
import pandas as pd
import time
```

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
# observaciones pasadas  (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
# tiempos futuros  (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    agg = concat(cols, axis=1)
    agg.columns = names
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
data = pd.read_excel('../PreparacionDatos/FloraTotal.xlsx' , sheet_name='FloraTotal' ,header=0)
data = data.drop(data.columns[[0]], axis=1) 
```

```python
#Se dividen las columanas para entradas y salidas del modelo.
data_in=data[['DirVien','H2S','Humedad','Lluvia','RSolar','Tasa_crecimiento_general','Temp','VelVien']]
data_out=data[['CO','SO2','O3','NO2','PM10']]

#Saco los valores 
values_in = data_in.values
values_out = data_out.values

```

```python
# integer encode direction
encoder = LabelEncoder()
values[:,4] = encoder.fit_transform(values_in[:,4])
values_out = values_out.astype('float32')
```

```python
# normalize features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_in = scaler.fit_transform(values_in)
scaled_out = scaler.fit_transform(values_out)
```

```python
# Numero de horas
n_hours = 12
n_features_in = 8
n_features_out = 5


reframed_in = series_to_supervised(scaled_in, n_hours, 1)
reframed_out = series_to_supervised(scaled_out, n_hours, 1)
print(reframed_in.head())
print(reframed_out.head())
```

```python
values_in = reframed_in.values
values_out = reframed_out.values

#Un ao de datos
n_train_hours = 365 * 24

#Valores de entrada y salida

train_in = values_in[:n_train_hours, :]
train_out = values_out[:n_train_hours, :]

test_in = values_in[n_train_hours:, :]
test_out = values_out[n_train_hours:, :]


# Caracteristicas de entrada y salida.
n_obs_in = n_hours * n_features_in
n_obs_out = n_hours * n_features_out
```

```python
#Datos de entramiento para la entrada y salida.
train_X_in, train_y_in = train_in[:, :n_obs_in], train_in[:, -n_features_in]
test_X_in, test_y_in, = test_in[:, :n_obs_in], test_in[:, -n_features_in]
```

```python
#Datos de prueba para la entrada y salida.
train_X_out, train_y_out = train_out[:, :n_obs_out], train_out[:, -n_features_out]
test_X_out, test_y_out = test_out[:, :n_obs_in], test_out[:, -n_features_out]
```

```python
# Datos x para entreamiento en 3D [samples, timesteps, features]
train_X_in = train_X_in.reshape((train_X_in.shape[0], n_hours, n_features_in))
train_X_out = train_X_out.reshape((train_X_out.shape[0], n_hours, n_features_out))

# Datos y para entreamiento en 3D [samples, timesteps, features]
train_y_in = train_y_in.reshape((train_y_in.shape[0], n_hours, n_features_in))
train_Y_out = train_Y_out.reshape((train_Y_out.shape[0], n_hours, n_features_out))


# Datos  x pruebas en 3D [samples, timesteps, features]
test_X_in = test_X_in.reshape((test_X_in.shape[0], n_hours, n_features_in))
test_X_out = test_X_out.reshape((test_X_out.shape[0], n_hours, n_features_out))


# Datos y pruebas en 3D [samples, timesteps, features]
test_Y_in = test_Y_in.reshape((test_Y_in.shape[0], n_hours, n_features_in))
test_Y_out = test_X_out.reshape((test_Y_out.shape[0], n_hours, n_features_out))


print("Datos de entramiento"+ train_X_in.shape, train_X_out.shape, train_y_in.shape, train_Y_out.shape)
print("Datos de prueba"+test_X_in.shape, test_X_out.shape, test_Y_in.shape, test_Y_out.shape)
```

```python
#Diseando la red neuronal.
model = Sequential()
#Agrego una una LSTM(unidades, 1, 8)
model.add(LSTM(50, input_shape=(train_X_in.shape[1], train_X_in.shape[2])))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')

#fit(x=None, y=None, batch_size=None, epochs=1, 
#verbose=1, callbacks=None, validation_split=0.0, validation_data=None, 
#shuffle=True, class_weight=None, sample_weight=None, initial_epoch=0, steps_per_epoch=None,
#validation_steps=None, validation_freq=1, max_queue_size=10, workers=1, use_multiprocessing=False)


# Ajustado el modelo
history = model.fit(x=(train_X_in,train_X_out) y=(train_y_in,train_Y_out), epochs=50, batch_size=72,validation_data=((test_X_in,test_X_out), (test_Y_in,test_Y_out)), verbose=2, shuffle=False)

# Seguimiento al modelo 
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

```python
import numpy as np
from sklearn import metrics
from sklearn import datasets


def rmse(objectivo, estimaciones):
    return np.sqrt(metrics.mean_squared_error(objetivo, estimaciones)
                  )

def  ajustested_r2(objectivo, estimaciones,n,k):
    r2=metrics.r2_score(objetico,estimaciones)
    return 1-(1-r2)*(n-1)/(n-k-1)

def evaluar_modelo(objectivo,estimaciones,n,k):
    return {
        "rmse":rmse(objectivo,estimaciones),
        "mae": metrics.mean_absolute_error(objectivo,estimaciones),
        "ajustested_r2": ajustested_r2(objectivo,estimaciones,n,k)
    }
```
